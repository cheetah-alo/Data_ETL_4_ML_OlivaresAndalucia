# Dataset final

El objetivo es mergear el dataset de clima con el dataset de raif_muestras + parcelas

```python
import pandas as pd
clima_sort = pd.read_csv("clima_sort.csv",index_col=0)
clima_sort
```

```python
clima_sort['FECHA'] =  pd.to_datetime(clima_sort['FECHA'])
```

```python
clima_sort.columns
```

En un principio se eliminan estas columnas porque se tenia otro objetivo. Al final del NB se vuelven a añadir

```python
df_copy = clima_sort.copy()
df_copy = df_copy.drop(columns = ['DAY', 'MONTH', 'YEAR','DIRECCION', 'NOMBRE',
       'PROVINCIA', 'MUNICIPIO', 'LATITUD', 'LONGITUD', 'ALTITUD'])
```

```python
df_copy
```

```python
df_copy = df_copy.drop_duplicates(subset = ["FECHA","COD_EST"], keep = "first").reset_index(drop=True)
df_copy
```

Se agrupa por COD_EST y FECHA

```python
df_group = df_copy.groupby(['COD_EST','FECHA']).mean()
df_group
```

```python
df_group.head(20)
```

Para cada una de las columnas del dataframe:
- Se calcula pora cada muestra la media de los últimos 7 días

A continuación, se muestra un ejemplo para la columna CV4

```python
df_test = pd.DataFrame(df_copy.groupby("COD_EST")['CV4'].rolling(window=7).mean())
```

```python
df_test.head(20)
```

Se da formato de dataframe

```python
df_vero = df_test.reset_index(level=[0,1]).drop(columns = ["level_1"])
df_vero
```

Se llevan los valores al dataframe principal

```python
df_group['CV4'] = df_vero['CV4'].values
df_group
```

Se vuelve a formatear el DF

```python
df_group = df_group.reset_index(level=[0,1])
df_group
```

```python
df_group.to_csv("clima_group.csv")
```
