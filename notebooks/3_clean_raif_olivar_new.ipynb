Se verifica si una estación climática asociada no tiene valor pero su sensor climático asociado si tenía valor

```python
import numpy as np

def check_different_two_values_columns(columna1, columna2, df):
    count = 0
    for i in range(len(df[columna1])):
        if ((df.loc[i, columna1] == "") and (df.loc[i, columna2] != "")):
            count += 1
            print("columna1:", df.loc[i,columna1] , "columna2:", df.loc[i,columna2],"\n")
    print(count)
```

```python
check_different_two_values_columns("_x0034_01_x0020_Estación_x0020_climática_x0020_asociada","_x0034_02_x0020_Sensor_x0020_climático_x0020_asociado",parcela_cordoba)
```
