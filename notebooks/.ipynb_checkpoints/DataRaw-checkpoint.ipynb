```python
import numpy as np

def check_nan_two_columns(columna1, columna2, df):
    count = 0
    for i in range(len(df[columna1])):
        if(np.isnan(df.loc[i, columna1]) and (np.isnan(df.loc[i, columna2]))):
                count +=1
    return count
```

```python
(df_copy["V_MAX"].isna().sum() + df_copy["V_Max"].isna().sum()) - check_nan_two_columns("V_MAX","V_Max", df_copy)
```
